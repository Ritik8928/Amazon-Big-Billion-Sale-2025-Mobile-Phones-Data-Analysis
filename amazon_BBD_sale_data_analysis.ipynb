```python
df_clean = df.dropna(subset=["Price" , "Rating"]).copy()   #drop missing price/rating for some visualizations
```

```python
df["Brand"] = df["Product_Name"].str.split().str[0]    #extract brand name
df_clean["Brand"] = df["Product_Name"].str.split().str[0]
```

```python
# price distribution
plt.figure(figsize=(8,5))
sns.histplot(df_clean["Price"], bins=40, kde=True, color="blue")
plt.title("Price Distribution of Mobile Phones")
plt.xlabel("Price (INR)")
plt.ylabel("Count")
plt.show()
```

```python
# Top Brands by Count of Phones
top_brands = df["Brand"].value_counts().head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_brands.values, y=top_brands.index, palette="viridis", hue = True)
plt.title("Top 10 Brands by Number of Phones")
plt.xlabel("Number of Phones")
plt.ylabel("Brand")
plt.show()
```

```python
# Average Rating per Brand (Top 10 by phone count)
avg_rating = df.groupby("Brand")["Rating"].mean().dropna().sort_values(ascending=False)
top_avg_rating = avg_rating.head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_avg_rating.values, y=top_avg_rating.index, palette="coolwarm")
plt.title("Top 10 Brands by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Brand")
plt.show()
```

```python
#Price vs Rating
plt.figure(figsize=(8,5))
sns.scatterplot(data=df_clean, x="Price", y="Rating", hue="Brand", alpha=0.6, legend=False)
plt.title("Price vs Rating of Mobiles")
plt.xlabel("Price (INR)")
plt.ylabel("Rating")
plt.show()
```

```python
#Review Count vs Rating
plt.figure(figsize=(8,5))
sns.scatterplot(data=df_clean, x="Review_Count", y="Rating", alpha=0.6, color="purple")
plt.title("Review Count vs Rating")
plt.xlabel("Number of Reviews")
plt.ylabel("Rating")
plt.show()
```

```python
# Correlation Heatmap (Price, Rating, Review_Count)
plt.figure(figsize=(6,4))
sns.heatmap(df_clean[["Price", "Rating", "Review_Count"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Boxplot Price Distribution by Top 5 Brands
top5_brands = df["Brand"].value_counts().head(5).index
plt.figure(figsize=(10,6))
sns.boxplot(data=df_clean[df_clean["Brand"].isin(top5_brands)], x="Brand", y="Price", palette="Set2")
plt.title("Price Distribution by Top 5 Brands")
plt.xlabel("Brand")
plt.ylabel("Price (INR)")
plt.show()
```

```python
# Regression Plot Price vs Rating (Trend line)
plt.figure(figsize=(8,5))
sns.regplot(data=df_clean, x="Price", y="Rating", scatter_kws={'alpha':0.3}, line_kws={"color":"red"})
plt.title("Price vs Rating with Trend Line")
plt.xlabel("Price (INR)")
plt.ylabel("Rating")
plt.show()
```
